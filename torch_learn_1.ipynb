```python
from torch.autograd import Variable
import torch
import numpy as np
import matplotlib.pyplot as plt


x= np.arange(-3,3.01,0.1)
y = x*x
plt.plot(x,y)
plt.plot(2,4,'ro')
plt.show()

x = Variable(torch.FloatTensor([2]),requires_grad=True)
# y = Variable(torch.tensor(y))
y = x**2
y.backward()
print(x.grad)
```
